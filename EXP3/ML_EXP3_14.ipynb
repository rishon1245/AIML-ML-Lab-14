```python
# Step 1: Imports
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error,r2_score
```

```python
# Step 2: Input data
X = np.array([10,20,30,50]).reshape(-1,1)
y = np.array([12,21,29,48])
```

```python
# Step 3: Create model and fit data
model = LinearRegression()
model.fit(X,y)
```

```python
# Step 4: Get coefficents and print
w = model.coef_[0]
b = model.intercept_
print(f"The slope is : {w}\nThe intercept is : {b}")
```

```python
# Step 5: Make predictions
y_pred = model.predict(X)
print("Predictions for training data\n")
for xi, yi, ypi in zip(X.flatten(), y, y_pred):
    print(f"X = {xi}, Actual y = {yi}, Predicted y = {ypi}")
```

```python
# Step 6: Error Calculation
mse = mean_squared_error(y, y_pred)
r2 = r2_score(y, y_pred)
print(f"Mean Squared Error is: {mse}")
print(f"R2_Score is: {r2}")
```

<h2>Linear Regression using MLE</h2>

```python
# Step 1: Imports
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error,r2_score
from scipy.optimize import minimize
```

```python
# Step 2: Input data
X = np.array([1,2,3])
y = np.array([2,3,5])
```

```python
# Step 3: Negative Log Likelihood function
def neg_log_likelihood(parms):
    w, b = parms
    sigma2 = 1  #assume varience = 1
    y_pred = w*X + b
    nll = 0.5*np.sum((y-y_pred)**2 / sigma2)
    return nll
```

```python
#Initial values for w (weight) and b (bias)
initial_guess = [0,0]
```

```python
# Step 4: Minimize nll
result = minimize(neg_log_likelihood, initial_guess)
w_mle, b_mle = result.x
print(f"Slope is: {w_mle}\nIntercept is: {b_mle}")
```

```python
# Step 5: Prediction
y_pred = w_mle*X + b_mle
print("Predictions for training data\n")
for xi, yi, ypi in zip(X.flatten(), y, y_pred):
    print(f"X = {xi}, Actual y = {yi}, Predicted y = {ypi}")
```

```python
# Step 6: Error Calculation
mse = mean_squared_error(y, y_pred)
r2 = r2_score(y, y_pred)
print(f"Mean Squared Error is: {mse}")
print(f"R2_Score is: {r2}")
```

<h2>Linear Regresssion using Gradient Descent</h2>

```python
# Step 1: Imports
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error,r2_score
```

```python
# Step 2: Input data
X = np.array([1,2,3])
y = np.array([2,3,5])
```

```python
# Step 3:
w, b = 0, 0
alpha = 0.001
n_iter = 100000
n = len(X)
```

```python
for i in range(n_iter):
    y_pred = w*X + b
    dw = (-2/n)*np.sum(X*(y-y_pred))
    db = (-2/n)*np.sum(y-y_pred)
    w -= alpha*dw
    b -= alpha*db

print(f"Slope is: {w}\nIntercept is: {b}")
```

<h2>Linear Regresssion using Gradient Descent with single parameter</h2>

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Step 2: Input data
X = np.array([1,2,3])
y = np.array([2,3,5])
n = len(X)
```

```python
# Step 3: Calculate loss function
def loss(w1):
    w0 = np.mean(y) - w1*np.mean(X)
    y_pred = w1*X + w0
    return np.sum((y-y_pred)**2)
```

```python
# Step 4: Calculate Gradient of J wrt w1
def gradient(w1):
    w0 = np.mean(y) - w1*np.mean(X)
    y_pred = w1*X + w0
    return (-2/n)*np.sum(X*(y-y_pred))
```

```python
# Step 5: Gradient Descent Algorithm 
lr = 0.1
w1 = 4
n_iter = 15
w1_values = []
loss_values = []

for i in range(n_iter):
    w1_values.append(w1)
    loss_values.append(loss(w1))
    grad = gradient(w1)
    w1 -= lr*grad
```

```python
# Step 6: plot Loss function and GD
w_space = np.linspace(-5,5,200)
loss_space = [loss(w) for w in w_space]
plt.figure(figsize = (7,7))
plt.plot(w_space,loss_space, label="J(w1)")
plt.scatter(w1_values,loss_values,color="red",label="GD Steps")
plt.xlabel("Slope (w1)")
plt.ylabel("Loss (J)")
plt.title("Gradient Descent using Slope variable")
plt.legend()
plt.grid(True)
plt.show()
```
